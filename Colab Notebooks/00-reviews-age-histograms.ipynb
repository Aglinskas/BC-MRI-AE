```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df_abide = pd.read_csv('../Data/ABIDE_legend_S982_v2.csv')
df_sfari = pd.read_csv('../Data/sfari_legend_S121.csv')
```

```python
plt.hist(df_abide['AgeAtScan'].values, bins=20, rwidth=0.9,color='#607c8e');
plt.title('ABIDE dataset')
plt.ylabel('Count')
plt.xlabel('Age (years)')
```

```python
df_sfari_16p = df_sfari[df_sfari['family_type'].isin(['16p-duplication','16p-deletion'])]
```

```python
plt.hist(df_sfari_16p['age_months'].values/12,bins=20, rwidth=0.9,color='#607c8e');
plt.title('SFARI dataset')
plt.ylabel('Count')
plt.xlabel('Age (years)')
```

```python
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.hist(df_abide['AgeAtScan'].values, bins=20, rwidth=0.9,color='#607c8e');
plt.title('ABIDE dataset')
plt.ylabel('Count')
plt.xlabel('Age (years)')

plt.subplot(1,2,2)
plt.hist(df_sfari_16p['age_months'].values/12,bins=20, rwidth=0.9,color='#607c8e');
plt.title('SFARI dataset')
plt.ylabel('Count')
plt.xlabel('Age (years)')

plt.suptitle('Subject Ages',fontsize=16,y=1.02)
```

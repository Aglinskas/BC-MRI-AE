```python
%%capture
keys = ['ADOS_Comm', 'ADOS_Social', 'ADOS_StBeh','ADOS_Total','FIQ','AgeAtScan','PIQ','VIQ','ADI_R_RRB','ADI_R_SocialTotal','ADI_R_VerbalTotal','Handedness_Scores']
res = np.zeros((2,len(keys),10))
for i in range(10):
    temp_umap = umap_data['reducer_sl_abide'].transform(data['salient_vec_abide'][i,patients,:])
    dim1 = temp_umap[:,0];
    dim2 = temp_umap[:,1];
    for k in range(len(keys)):
        vec = df[keys[k]].values[patients];
        v = ~np.isnan(vec);
        res[0,k,i] = np.corrcoef(dim1[v],vec[v])[0,1];
        res[1,k,i] = np.corrcoef(dim2[v],vec[v])[0,1];
        
plt.figure(figsize=(10,5));
xs = np.linspace(0,res.shape[1]*2,res.shape[1]);
plt.bar(xs-.5,res[0,:,:].mean(axis=1));
plt.errorbar(xs-.5,res[0,:,:].mean(axis=1),res[0,:,:].std(axis=1),fmt='r ');

plt.bar(xs+.5,res[1,:,:].mean(axis=1));
plt.errorbar(xs+.5,res[1,:,:].mean(axis=1),res[1,:,:].std(axis=1),fmt='r ');

plt.xticks(xs,labels=keys,rotation=45);
```

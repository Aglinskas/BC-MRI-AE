```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv('/Users/aidasaglinskas/Desktop/BC-MRI-AE/CSVs/ABIDE-legend.csv')
df2 = pd.read_csv('/Users/aidasaglinskas/Downloads/ABIDEII_Composite_Phenotypic.csv')
```

```python
df1 = df1[df1['DxGroup']==1]
df2 = df2[df2['DX_GROUP']==1]
```

```python
perc = list()
str_list = list()
for col in list(df2.columns):
    p = (~pd.isna(df2[col]).values).mean()*100
    perc.append(p)
    str_list.append(f'{col} : {p :.2f}%')
    
list(np.array(str_list)[np.argsort(perc)[-1:0:-1]]) 
```
